```python
import numpy as np
import matplotlib.pyplot as plt
from dipy.io.image import load_nifti, save_nifti
```

```python
hcp_7T_lev_map, affine =  load_nifti('lev_map_7T.nii.gz')
sh_lev_map, affine = load_nifti('lev_map_StanfordHARDI.nii.gz')
ppmi_lev_map, affine = load_nifti('lev_map.nii.gz')
```

```python
orig = hcp_7T_lev_map[:, :, 75]
den = ppmi_lev_map[:, :, 40]
rms_diff = sh_lev_map[:, :, 40]

fig1, ax = plt.subplots(1, 3, figsize=(12, 6),
                        subplot_kw={'xticks': [], 'yticks': []})

fig1.subplots_adjust(hspace=0.3, wspace=0.05)

ax.flat[0].imshow(orig.T, cmap='magma', interpolation='none',
                  origin='lower')
ax.flat[0].set_title('Original')
ax.flat[1].imshow(den.T, cmap='magma', interpolation='none',
                  origin='lower')
ax.flat[1].set_title('Denoised Output')
ax.flat[2].imshow(rms_diff.T, cmap='magma', interpolation='none',
                  origin='lower')
ax.flat[2].set_title('Residuals')

fig1.savefig('denoised_patch2self.png')

print("The result saved in denoised_patch2self.png")
```
