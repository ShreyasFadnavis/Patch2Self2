```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('total_speedups.csv')
ax = sns.pointplot(x="Sketch Size", y="Speedup", data=df,  hue="Dataset", size_order='Dataset', markers=['o', 'x', 'D'],
                  linestyles=['-', '--', '-.'])
plt.savefig('sim_p2s.png', dpi=1000)
```

```python
import seaborn as sns
import pandas as pd
df = pd.read_csv('time_sim_p2s.csv')
ax = sns.pointplot(x="Sketch Size", y="Time", data=df,  hue="Dataset", palette="Set2")
plt.axhline(y=69.86860918998718, color='r', linestyle='-')
ax.get_xaxis().set_ticks([])
plt.savefig('sim_p2s.png', dpi=1000)
```
