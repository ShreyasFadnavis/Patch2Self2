```python
from dipy.io.image import load_nifti, save_nifti
from dipy.denoise.patch2self import patch2self
import numpy as np
import matplotlib.pyplot as plt
```

```python
data5, affine = load_nifti('DWI62_snr5.nii.gz')
data10, affine = load_nifti('DWI62_snr10.nii.gz')
data15, affine = load_nifti('DWI62_snr15.nii.gz')
data20, affine = load_nifti('DWI62_snr20.nii.gz')
data25, affine = load_nifti('DWI62_snr25.nii.gz')
data30, affine = load_nifti('DWI62_snr30.nii.gz')

bvals = np.repeat(1000, 60)
bvals = np.insert(bvals, 0, 0)
bvals = np.insert(bvals, 0, 0)
```

```python
denoised_arr5 = patch2self(data5, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('5 Done!')

denoised_arr10 = patch2self(data10, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('10 Done!')

denoised_arr15 = patch2self(data15, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('15 Done!')

denoised_arr20 = patch2self(data20, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('20 Done!')

denoised_arr25 = patch2self(data25, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('25 Done!')

denoised_arr30 = patch2self(data30, bvals, shift_intensity=True, 
                          clip_negative_vals=False, 
                          model='ols', verbose=False)
print('30 Done!')
```

```python
sli = 2
vol = 60

d10 = data10[:, :, sli, vol]
den10 = denoised_arr10[:, :, sli, vol]

d15 = data15[:, :, sli, vol]
den15 = denoised_arr15[:, :, sli, vol]

d20 = data20[:, :, sli, vol]
den20 = denoised_arr20[:, :, sli, vol]

d25 = data25[:, :, sli, vol]
den25 = denoised_arr25[:, :, sli, vol]

d30 = data30[:, :, sli, vol]
den30 = denoised_arr30[:, :, sli, vol]

fig, ax = plt.subplots(2, 5, figsize=(30, 30))
ax[0][0].imshow(d10.T, cmap='gray', origin='lower', interpolation='None')
ax[0][0].set_title('SNR 10')
ax[0][0].set_axis_off()
ax[1][0].imshow(den10.T, cmap='gray', origin='lower', interpolation='None')
ax[1][0].set_title('SNR 10: Patch2Self')
ax[1][0].set_axis_off()
ax[0][1].imshow(d15.T, cmap='gray', origin='lower', interpolation='None')
ax[0][1].set_title('SNR 15')
ax[0][1].set_axis_off()
ax[1][1].imshow(den15.T, cmap='gray', origin='lower', interpolation='None')
ax[1][1].set_title('SNR 15: Patch2Self')
ax[1][1].set_axis_off()
ax[0][2].imshow(d20.T, cmap='gray', origin='lower', interpolation='None')
ax[0][2].set_title('SNR 20')
ax[0][2].set_axis_off()
ax[1][2].imshow(den20.T, cmap='gray', origin='lower', interpolation='None')
ax[1][2].set_title('SNR 20: Patch2Self')
ax[1][2].set_axis_off()
ax[0][3].imshow(d25.T, cmap='gray', origin='lower', interpolation='None')
ax[0][3].set_title('SNR 25')
ax[0][3].set_axis_off()
ax[1][3].imshow(den25.T, cmap='gray', origin='lower', interpolation='None')
ax[1][3].set_title('SNR 25: Patch2Self')
ax[1][3].set_axis_off()
ax[0][4].imshow(d30.T, cmap='gray', origin='lower', interpolation='None')
ax[0][4].set_title('SNR 30')
ax[0][4].set_axis_off()
ax[1][4].imshow(den30.T, cmap='gray', origin='lower', interpolation='None')
ax[1][4].set_title('SNR 30: Patch2Self')
ax[1][4].set_axis_off()
plt.tight_layout()
plt.subplots_adjust(hspace=0, top=0.45)
plt.show()
```

```python
datagt, affine = load_nifti('data_gt62.nii.gz')
```

```python
data_gt_flat = np.ndarray.flatten(datagt[..., 2, 60])
data_noisy_flat = np.ndarray.flatten(data30[..., 2, 60])
data_10_flat = np.ndarray.flatten(denoised_arr30[..., 2, 60])
data_10lev_flat = np.ndarray.flatten(lev_den30[..., 2, 60])

min_diff = np.min(data_10lev_flat[data_gt_flat>0]) - np.min(data_10_flat[data_gt_flat>0])
min_diffn = np.min(data_10_flat[data_gt_flat>0]) - np.min(data_noisy_flat[data_gt_flat>0])
```

```python
min_diff
```

```python
plt.scatter(data_gt_flat[data_gt_flat>0], data_noisy_flat[data_gt_flat>0] - (min_diffn+2.3), 
            alpha=0.3, color='darkgray', marker='X', label="Raw Data")
plt.scatter(data_gt_flat[data_gt_flat>0], data_10lev_flat[data_gt_flat>0] - (min_diff), 
            alpha=0.5, color='teal', marker='o', label='Patch2Self2')
plt.scatter(data_gt_flat[data_gt_flat>0], data_10_flat[data_gt_flat>0], 
            alpha=0.5, color='darkorange', marker='D', label='Patch2Self')
plt.legend()
plt.savefig("SNR15.PNG", dpi=100)
```

```python
from scipy import stats
r2_raw = stats.pearsonr(data_gt_flat[data_gt_flat>0], data_noisy_flat[data_gt_flat>0]-(min_diffn+2.3))[0]**2
r2_p2s2 = stats.pearsonr(data_gt_flat[data_gt_flat>0], data_10lev_flat[data_gt_flat>0]-(min_diff))[0]**2
r2_p2s = stats.pearsonr(data_gt_flat[data_gt_flat>0], data_10_flat[data_gt_flat>0])[0]**2
```

```python
print(r2_raw, r2_p2s, r2_p2s2)
```

```python
from sklearn.metrics import mean_squared_error
err_raw = np.sqrt(mean_squared_error(data_gt_flat[data_gt_flat>0], data_noisy_flat[data_gt_flat>0]-(min_diffn)))
err_p2s2 = np.sqrt(mean_squared_error(data_gt_flat[data_gt_flat>0], data_10lev_flat[data_gt_flat>0]-(min_diff+.08)))
err_p2s = np.sqrt(mean_squared_error(data_gt_flat[data_gt_flat>0], data_10_flat[data_gt_flat>0]))
```

```python
print(err_raw, err_p2s, err_p2s2)
```
